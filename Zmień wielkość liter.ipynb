```python
letters = "aBDc"

def swap_cases(letters=letters):
    ### wpisz kod poniżej
    result = ""
    for letter in letters:
      if letter.islower():
        result += letter.upper()
      elif letter.isupper():
        result += letter.lower()
      elif not letter.isalpha():
        result += letter

    #### wpisz kod powyżej
    return result


print(swap_cases(letters))
```
