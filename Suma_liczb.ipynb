```python
elements = [1, " ", 8.5, "blabla", 5]

def find_sum(elements=elements):
    ### wpisz kod poniżej
    result = 0
    for el in elements:
        if isinstance(el, (int, float)):
            result += el

    #### wpisz kod powyżej
    return result


print(find_sum(elements))
```
