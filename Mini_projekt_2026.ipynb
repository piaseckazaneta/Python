Funkcja zamieniająca tekstowe wartości zniszczeń (np. 10.00K, 2.50M) na czyste liczby typu float ($10000$, $2500000$) dla kolumn "damage_crops" i "damage_property" w df_details_analysis

```python
df_details_analysis["damage_property"].unique()
```

```python
def damage_costs(damage_value):
  if pd.isna(damage_value):
    return 0
  elif "K" in damage_value:
    damage_value = damage_value[:-1]
    damage_value = float(damage_value) * 1000
    return damage_value
  elif "M" in damage_value:
    damage_value = damage_value[:-1]
    damage_value = float(damage_value) * 1000000
    return damage_value
  elif "B" in damage_value:
    damage_value = damage_value[:-1]
    damage_value = float(damage_value) * 1000000000
    return damage_value
  else:
    float(damage_value)
```

```python
df_details_analysis["damage_property"] = df_details_analysis["damage_property"].apply(damage_costs)
```

```python
print(df_details_analysis["damage_property"])
```

```python
df_details_analysis["damage_crops"] = df_details_analysis["damage_crops"].apply(damage_costs)
```

```python
print(df_details_analysis["damage_crops"])
```

Wyciągnięcie z kolumny "tor_f_scale" samej liczby (wartość wyrażona jest w skali Fujita, np. EF1). Wartość liczbowa będzie niezbędna do obliczeń zależności powierzchni zniszczeń od wzrostu siły tornada.

Sprawdzenie unikalnych wartości występujących w kolumnie "tor_f_scale"

Funkcja wyodrębniająca cyfrę z wartości w kolumnie "tor_f_scale".
'EFU' jest oznaczeniem tornada, które przechodziło przez pole i nie spowodowało uszkodzeń budynków, zatem nie można mu przypisać oznaczenia w skali 1-5.

```python
def scale_number(scale):
  if pd.isna(scale):
    return 'NaN'
  elif "EFU" in scale:
    return 'NaN'
  elif "EF" in scale:
    scale = scale[2:]
    return scale
  else:
    return scale
```

Zastosowanie napisanej funkcji dla kolumny "tor_f_scale" i utworzenie nowej kolumny z samą cyfrą.

```python
df_details_analysis["tor_f_scale_number"] = df_details_analysis["tor_f_scale"].apply(scale_number)
```

Sprawdzenie unikalnych wartości występujących w nowej kolumnie

```python
df_details_analysis["tor_f_scale_number"].unique()
```

Zamiana wartości na float.

```python
df_details_analysis["tor_f_scale_number"] = (df_details_analysis["tor_f_scale_number"]).astype(float)
```
